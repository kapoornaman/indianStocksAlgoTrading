# Creating our Dataframe

```python
my_columns=["Ticker","Price","Market Capitalization(Trillions)","Number of Shares to Buy"]
    
```

```python
final_dataframe = pd.DataFrame(columns = my_columns)

    
```

# Looping Through The Tickers in Our List of Stocks and Adding them in our Dataframe

```python
for ticker in df['Symbol']:
    url=f"https://www.google.com/finance/quote/{ticker}:NSE?hl=en"
    response=requests.get(url)
    soup=BeautifulSoup(response.text,'html.parser')
    price=float(soup.find(class_=class1).text.strip()[1:].replace(',', ''))
    marketcap=soup.find_all(class_=class2)
    marketcap=float(marketcap[3].text.strip()[:4])
    if(marketcap>100):
        marketcap/=100
    final_dataframe.loc[len(final_dataframe)] = [ticker,
                                                   price, 
                                                marketcap, 
                                                   'N/A']
final_dataframe
```

# Taking The Porfolio Size As Input And Recommending As Per Portfolio

```python
portfolio_size = input("Enter the value of your portfolio:")
```

## Caluclating The Number Of Shares To Buy

```python
position_size = float(portfolio_size) / len(final_dataframe.index)
position_size
for i in range(0, len(final_dataframe['Ticker'])):
    final_dataframe.loc[i, 'Number of Shares to Buy'] = math.floor(position_size / final_dataframe['Price'][i])
```
